### Simulating trend data for the number of customers 

```python
import numpy as np

# Parameters
initial_customers = 1000  # Starting number of customers
growth_rate = 0.2  # 20% growth rate (adjust as needed)
years = np.arange(1, 6)  # 5 years

# Simulate customer growth using exponential function
customers = initial_customers * np.exp(growth_rate * years)

# Print the simulated customer data
for year, num_customers in zip(years, customers):
    print(f"Year {year}: {num_customers:.0f} customers")
```

## Financial Modelling

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
M = 2500  # Price of Speak Toons
C = 100000  # Cost to develop and maintain your product
years = np.arange(1, 6)  # 5 years

# Simulated market trend data (number of customers)

customers = np.array([1221, 1492,1822, 2226, 2718])

# Calculate total profit (Y) based on the equation Y = M * X(t) - C
total_profit = M * customers - C

# Create the graph
plt.figure(figsize=(10, 6))
plt.plot(years, total_profit, marker='o')
plt.xlabel('Years')
plt.ylabel('Total Profit')
plt.title('Total Profit vs. Years')
plt.xticks(ticks=years, labels=['Year 1', 'Year 2', 'Year 3', 'Year 4', 'Year 5'])
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [1, 2, 3, 4, 5]
revenues = [2500 * customers[i] for i in range(len(years))]

plt.plot(years, revenues, marker='o', color='orange')
plt.xlabel('Years')
plt.ylabel('Revenue')
plt.title('Revenue Projection')
plt.grid(True)
plt.show()
```
